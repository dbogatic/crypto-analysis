# Pytrends 
## Allows simple interface for automating downloading of reports from Google Trends
https://pypi.org/project/pytrends/

```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
```

# ETH Google Trends

```python
# Execute the TrendReq method by passing the host language (hl) and timezone (tz) parameters
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
# Build list of keywords
kw_list = ["ethereum"] 

# Build the payload
eth_coin_trends = pytrends.build_payload(kw_list, timeframe='2021-01-01 2021-07-12')
```

```python
# Pull ETH Google trends daily data for the selected time period
eth_trends = pytrends.interest_over_time()

eth_trends.to_csv('resources/eth_google_trends/eth_google_trends.csv')
```

```python
eth_metrics = pd.read_csv('/Users/draganbogatic/Documents/Fin_Tech/crypto-analysis/resources/eth_annual_metrics/eth_metrics_1.csv', parse_dates=True, index_col="time")
```

```python
# Pull ETH prices from csv file
eth_price = eth_metrics['PriceUSD']
eth_price.head()
```

```python
# Plot ETH price vs Google trends
fig, ax = plt.subplots(figsize=(12,7))
plt.plot(eth_metrics['PriceUSD'], label ='ETH Price', color='r')
ax.set_ylabel('ETH Price ($)')
plt.xticks(rotation=45)
ax.legend(loc='upper left')
plt.grid()

ax2 = ax.twinx()
plt.plot(eth_trends['ethereum'], label='ETH Google Trends', color='b')
ax.set_title('ETH Price vs Google Trends')
ax2.set_ylabel('ETH Google Trends') 
ax2.legend(loc='upper right')
```
